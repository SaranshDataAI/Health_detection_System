```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('G:\VSCODE\Python And Data Sceince\Advance AI Projects\Health_care_system\Data\Healthcare.csv')
```

```python
print("Shape:", df.shape)
print("\nColumns and dtypes:")
print(df.dtypes)
print("\nFirst 5 rows:")
display(df.head())

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nUnique diseases count:", df['Disease'].nunique())
print("\nTop 20 diseases by frequency:")
print(df['Disease'].value_counts().head(20))

# show basic age distribution and gender distribution
print("\nAge describe:")
print(df['Age'].describe())

print("\nGender value counts:")
print(df['Gender'].value_counts())
```

```python
# Cell 3 — Clean Symptoms column and extract symptom vocabulary

def clean_symptom_list(sym_str):
    """Convert 'fever, nausea, cough' → ['fever','nausea','cough']"""
    return [s.strip().lower() for s in str(sym_str).split(',') if s.strip()]

# Apply cleaning
df['Clean_Symptoms'] = df['Symptoms'].apply(clean_symptom_list)

# Build full symptom vocabulary
from collections import Counter

symptom_counter = Counter()

for symptoms in df['Clean_Symptoms']:
    symptom_counter.update(symptoms)

print("Total unique symptoms:", len(symptom_counter))
print("\nTop 50 most common symptoms:")
for sym, cnt in symptom_counter.most_common(50):
    print(f"{sym:25s}  {cnt}")
```

```python
# Cell 4 — Multi-Hot Encode Symptoms

# Create a sorted list of all symptoms for consistent column order
all_symptoms = sorted(symptom_counter.keys())

print("Symptoms being encoded:", all_symptoms, "\n")

# Create a column for each symptom (0/1)
for symptom in all_symptoms:
    df[f"sym_{symptom.replace(' ', '_')}"] = df['Clean_Symptoms'].apply(lambda x: 1 if symptom in x else 0)

# Check the first 5 rows with new symptom columns
print("New dataframe shape:", df.shape)
df.head()
```

```python
# Cell 5 — Label serious / long-term diseases

serious_keywords = [
    "heart", 
    "stroke",
    "diabetes",
    "hypertension",
    "liver",
    "kidney",
    "parkinson",
    "dementia",
    "cancer"
]

def is_serious_disease(disease_name):
    name = disease_name.lower()
    return any(keyword in name for keyword in serious_keywords)

df["Is_Serious"] = df["Disease"].apply(is_serious_disease)

print("Serious disease counts:")
print(df["Is_Serious"].value_counts())

print("\nSerious diseases identified:")
print(df[df["Is_Serious"] == True]["Disease"].unique())
```

```python
# Cell 6 — Prepare ML features and target

from sklearn.preprocessing import LabelEncoder

# Copy dataset to avoid modifying original
data = df.copy()

# Encode Gender (string → numeric)
gender_map = {"Male": 0, "Female": 1, "Other": 2}
data["Gender_Encoded"] = data["Gender"].map(gender_map)

# List of symptom columns
symptom_cols = [col for col in data.columns if col.startswith("sym_")]

# Final feature columns
feature_cols = ["Age", "Gender_Encoded", "Symptom_Count"] + symptom_cols + ["Is_Serious"]

print("Total feature columns:", len(feature_cols))
print(feature_cols)

# X and y
X = data[feature_cols]
y = data["Disease"]

# Encode disease labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("\nNumber of classes:", len(label_encoder.classes_))
print("Classes:", label_encoder.classes_)
```

```python
# Cell 7 — Train/Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y_encoded, 
    test_size=0.20, 
    random_state=42,
    stratify=y_encoded
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)

# Count distribution of classes in train/test
import numpy as np
unique, counts = np.unique(y_train, return_counts=True)
print("\nClass distribution in TRAIN:")
for u, c in zip(unique, counts):
    print(f"{label_encoder.classes_[u]:25s} {c}")

unique_t, counts_t = np.unique(y_test, return_counts=True)
print("\nClass distribution in TEST:")
for u, c in zip(unique_t, counts_t):
    print(f"{label_encoder.classes_[u]:25s} {c}")
```

```python
# Cell 8 — Compute Class Weights with Boost for Serious Diseases

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Get basic inverse-frequency class weights
class_weights_basic = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weights_basic = dict(enumerate(class_weights_basic))

# Add serious disease boost
serious_boost = 2.0   # You can increase to 3.0 later if needed

# Map serious diseases to their encoded label
serious_labels = [
    label_encoder.transform([d])[0]
    for d in ["Heart Disease", "Stroke", "Liver Disease", "Diabetes",
              "Hypertension", "Dementia", "Parkinson's", "Chronic Kidney Disease"]
]

class_weights_custom = {}

for label, base_w in class_weights_basic.items():
    if label in serious_labels:
        class_weights_custom[label] = base_w * serious_boost
    else:
        class_weights_custom[label] = base_w

print("=== Basic class weights ===")
for cls, weight in class_weights_basic.items():
    print(f"{label_encoder.classes_[cls]:25s}: {weight:.3f}")

print("\n=== Serious disease boosted weights ===")
for cls, weight in class_weights_custom.items():
    print(f"{label_encoder.classes_[cls]:25s}: {weight:.3f}")
```

```python
# Cell 9 — Train XGBoost classifier

from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score

# Build sample weights for each training row based on its class
sample_weights = np.array([class_weights_custom[label] for label in y_train])

model = XGBClassifier(
    objective='multi:softprob',
    num_class=len(label_encoder.classes_),
    eval_metric='mlogloss',
    learning_rate=0.08,
    max_depth=6,
    n_estimators=300,
    subsample=0.9,
    colsample_bytree=0.9,
    tree_method='hist',
    random_state=42
)

print("Training XGBoost model...")
model.fit(X_train, y_train, sample_weight=sample_weights)

print("Training complete.")

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate
acc = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", acc)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
# Cell 10 — TF-IDF + Logistic Regression (Much Higher Accuracy Expected)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Convert Clean_Symptoms list back to a string: "fever nausea cough"
df["Symptoms_Str"] = df["Clean_Symptoms"].apply(lambda x: " ".join(x))

# TF-IDF vectorizer (max 3000 features)
vectorizer = TfidfVectorizer(max_features=3000)
X_tfidf = vectorizer.fit_transform(df["Symptoms_Str"])

# Train/test split for TF-IDF features
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
    X_tfidf, 
    y_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_encoded
)

# Logistic Regression with class weights
log_reg = LogisticRegression(
    multi_class='ovr',
    class_weight=class_weights_custom,
    max_iter=500
)

print("Training Logistic Regression...")
log_reg.fit(X_train_t, y_train_t)

print("Training complete.")

# Prediction
y_pred_t = log_reg.predict(X_test_t)

# Evaluation
acc_t = accuracy_score(y_test_t, y_pred_t)
print("\nLogistic Regression Accuracy:", acc_t)

print("\nClassification Report:")
print(classification_report(y_test_t, y_pred_t, target_names=label_encoder.classes_))
```
