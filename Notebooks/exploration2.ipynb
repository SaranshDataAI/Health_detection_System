```python
import pandas as pd
import numpy as np
```

```python
df_main = pd.read_csv('G:\VSCODE\Python And Data Sceince\Advance AI Projects\Health_care_system\Data\dataset.csv')
df_desc = pd.read_csv('G:\VSCODE\Python And Data Sceince\Advance AI Projects\Health_care_system\Data\symptom_Description.csv')
df_prec = pd.read_csv('G:\VSCODE\Python And Data Sceince\Advance AI Projects\Health_care_system\Data\symptom_precaution.csv')
df_sev = pd.read_csv('G:\VSCODE\Python And Data Sceince\Advance AI Projects\Health_care_system\Data\Symptom-severity.csv')
```

```python
print("Main dataset shape:", df_main.shape)
print("Symptom description shape:", df_desc.shape)
print("Precautions shape:", df_prec.shape)
print("Symptom severity shape:", df_sev.shape)

print("\nMain Dataset Columns:")
print(df_main.columns)

print("\nFirst 5 rows of main dataset:")
display(df_main.head())
```

```python
# Cell 2 — Clean symptom columns and combine into a list

import numpy as np

# List all symptom columns
symptom_cols = [col for col in df_main.columns if "Symptom" in col]

print("Symptom columns found:", len(symptom_cols))

def clean_symptom(sym):
    """Clean symptom string: lowercase, strip spaces, remove weird characters."""
    if pd.isna(sym):
        return None
    return str(sym).strip().lower().replace(" ", "_")

# Create Clean_Symptoms list
df_main["Clean_Symptoms"] = df_main[symptom_cols].apply(
    lambda row: [clean_symptom(s) for s in row if pd.notna(s)],
    axis=1
)

print("\nSample cleaned symptoms:")
df_main[["Disease", "Clean_Symptoms"]].head(10)
```

```python
# Cell 3 — Prepare Symptom Severity Mapping

# Clean severity dataset
df_sev.columns = ["Symptom", "Severity"]
df_sev["Symptom"] = df_sev["Symptom"].str.strip().str.lower()

# Create mapping: symptom → severity
severity_map = dict(zip(df_sev["Symptom"], df_sev["Severity"]))

print("Total symptoms with severity mapping:", len(severity_map))
print("\nSample of severity map (first 15):")
list(severity_map.items())[:15]

# Function: compute severity score for each row
def compute_severity(symptom_list):
    return sum(severity_map.get(s, 0) for s in symptom_list)

df_main["Severity_Score"] = df_main["Clean_Symptoms"].apply(compute_severity)

# Additional helpful features
df_main["Symptom_Count"] = df_main["Clean_Symptoms"].apply(len)
df_main["Avg_Severity"] = df_main["Severity_Score"] / df_main["Symptom_Count"]

# Flag if severity is unusually high
df_main["High_Severity_Flag"] = df_main["Severity_Score"] > df_main["Severity_Score"].median()

print("\nPreview with severity added:")
df_main[["Disease", "Clean_Symptoms", "Severity_Score", "Symptom_Count", "Avg_Severity", "High_Severity_Flag"]].head(10)
```

```python
# Cell 4 — Multi-hot encode all symptoms

from collections import Counter

# Build symptom vocabulary
symptom_vocab = set()

for symptoms in df_main["Clean_Symptoms"]:
    symptom_vocab.update(symptoms)

symptom_vocab = sorted(list(symptom_vocab))

print("Total unique symptoms:", len(symptom_vocab))
print("Sample symptoms:", symptom_vocab[:20])

# Create 0/1 column for each symptom
for sym in symptom_vocab:
    df_main[f"sym_{sym}"] = df_main["Clean_Symptoms"].apply(lambda x: 1 if sym in x else 0)

print("\nNew dataframe shape:", df_main.shape)

# Show a preview of encoded symptoms
df_main.head()
```

```python
# Cell 5 — Label serious diseases

serious_keywords = [
    "heart", 
    "stroke",
    "diabetes",
    "hypertension",
    "paralysis",
    "hepatitis",
    "pneumonia",
    "chronic",
    "kidney",
    "cancer",
    "arthritis",  # (optional depending dataset)
    "brain",
    "alzheimer",
    "dementia",
    "attack"
]

def label_serious(disease):
    name = disease.lower()
    return any(word in name for word in serious_keywords)

df_main["Is_Serious"] = df_main["Disease"].apply(label_serious)

print("Serious disease counts:")
print(df_main["Is_Serious"].value_counts())

print("\nSerious diseases detected:")
print(df_main[df_main["Is_Serious"] == True]["Disease"].unique())
```

```python
# Cell 6 — Prepare ML features (X) and labels (y)

from sklearn.preprocessing import LabelEncoder

# Select symptom columns
symptom_feature_cols = [col for col in df_main.columns if col.startswith("sym_")]

# Other engineered features
engineered_features = ["Severity_Score", "Symptom_Count", "Avg_Severity", "High_Severity_Flag"]

# Final feature list
feature_cols = symptom_feature_cols + engineered_features

print("Total features used:", len(feature_cols))

# Prepare X matrix
X = df_main[feature_cols]

# Prepare label y
y = df_main["Disease"]

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Number of disease classes:", len(label_encoder.classes_))
print("Classes:", label_encoder.classes_)
```

```python
# Cell 7 — Stratified Train/Test Split

from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.20,
    random_state=42,
    stratify=y_encoded
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)

# Check class distribution
unique, counts = np.unique(y_train, return_counts=True)

print("\nClass distribution in TRAIN:")
for u, c in zip(unique, counts):
    print(f"{label_encoder.classes_[u]:35s}: {c}")

unique_t, counts_t = np.unique(y_test, return_counts=True)

print("\nClass distribution in TEST:")
for u, c in zip(unique_t, counts_t):
    print(f"{label_encoder.classes_[u]:35s}: {c}")
```

```python
# Cell 8 — Train Naive Bayes Classifier

from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, classification_report

nb_model = BernoulliNB()

print("Training Naive Bayes...")
nb_model.fit(X_train, y_train)

print("Training complete.")

# Predictions
y_pred_nb = nb_model.predict(X_test)

# Accuracy
nb_accuracy = accuracy_score(y_test, y_pred_nb)
print("\nNaive Bayes Accuracy:", nb_accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_nb, target_names=label_encoder.classes_))
```

```python
# Cell 9 — Save model and metadata

import pickle

# Save Naive Bayes model
with open("nb_model.pkl", "wb") as f:
    pickle.dump(nb_model, f)

# Save label encoder
with open("label_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder, f)

# Save symptom list (to build the UI dropdown in Flutter)
with open("symptom_list.pkl", "wb") as f:
    pickle.dump(symptom_vocab, f)

# Save severity map (for backend risk scoring)
with open("severity_map.pkl", "wb") as f:
    pickle.dump(severity_map, f)

print("All files saved successfully:")
print("- nb_model.pkl")
print("- label_encoder.pkl")
print("- symptom_list.pkl")
print("- severity_map.pkl")
```
